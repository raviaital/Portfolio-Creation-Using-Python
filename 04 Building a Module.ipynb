# Building a Module

Module in python is nothing but a python file. When you import a module the methods or variables in that file can be accessed.
But python imports the module only once by default. So, while developing it is also necessary to tell python to autoreload the module whenver there is a change. This can be done by loading the extension autoreload

%load_ext autoreload

%autoreload 2


```python
%load_ext autoreload
%autoreload 2
```
